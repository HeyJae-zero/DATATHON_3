```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 한글 폰트 설정
plt.rcParams['font.family'] = 'appleGothic'
plt.rcParams['axes.unicode_minus'] = False

# ========================================
# 1. 데이터 불러오기
# ========================================
df = pd.read_csv('/Users/h/Desktop/ms/vscode/my/dataton_project/data/online_shoppers/online_shoppers_intention.csv')

print("📊 데이터 기본 정보")
print(f"전체 세션 수: {len(df):,}개")
print(f"신규 방문자: {(df['VisitorType'] == 'New_Visitor').sum():,}개")
print(f"재방문자: {(df['VisitorType'] == 'Returning_Visitor').sum():,}개")

# ========================================
# 2. 신규 vs 재방문자 구분하기
# ========================================
# 신규 방문자만 추출
신규방문자 = df[df['VisitorType'] == 'New_Visitor']

# 재방문자만 추출
재방문자 = df[df['VisitorType'] == 'Returning_Visitor']

print(f"\n✅ 그룹 분류 완료!")
print(f"- 신규 방문자: {len(신규방문자):,}명")
print(f"- 재방문자: {len(재방문자):,}명")

# ========================================
# 3. 평균 이탈률 계산하기
# ========================================
# 신규 방문자의 평균 이탈률
신규_평균이탈률 = 신규방문자['BounceRates'].mean()

# 재방문자의 평균 이탈률
재방문_평균이탈률 = 재방문자['BounceRates'].mean()

print(f"\n\n{'='*50}")
print("📈 평균 이탈률 비교")
print(f"{'='*50}")
print(f"신규 방문자 평균 이탈률: {신규_평균이탈률:.4f} ({신규_평균이탈률*100:.2f}%)")
print(f"재방문자 평균 이탈률:   {재방문_평균이탈률:.4f} ({재방문_평균이탈률*100:.2f}%)")
print(f"\n차이: {신규_평균이탈률 - 재방문_평균이탈률:.4f} ({(신규_평균이탈률 - 재방문_평균이탈률)*100:.2f}%p)")

# ========================================
# 4. 가설 검정하기
# ========================================
print(f"\n\n{'='*50}")
print("🔍 가설 검정 결과")
print(f"{'='*50}")

# 재방문자가 더 낮은지 확인
if 재방문_평균이탈률 < 신규_평균이탈률:
    차이_배율 = 신규_평균이탈률 / 재방문_평균이탈률
    print(f"✅ 가설 충족!")
    print(f"   재방문자의 이탈률이 더 낮습니다!")
    print(f"   → 신규 방문자 대비 {차이_배율:.2f}배 낮은 이탈률")
else:
    print(f"❌ 가설 불충족!")
    print(f"   신규 방문자의 이탈률이 더 낮습니다.")

# ========================================
# 5. 통계적 유의성 검정 (t-test)
# ========================================
# 두 그룹의 이탈률 차이가 통계적으로 의미있는지 확인
t_stat, p_value = stats.ttest_ind(신규방문자['BounceRates'], 
                                   재방문자['BounceRates'])

print(f"\n\n📊 통계적 유의성 검정")
print(f"t-통계량: {t_stat:.4f}")
print(f"p-value: {p_value:.6f}")

if p_value < 0.05:
    print("→ ✅ 통계적으로 유의한 차이가 있습니다! (p < 0.05)")
    print("   이 차이는 우연이 아닙니다!")
else:
    print("→ ❌ 통계적으로 유의한 차이가 없습니다. (p ≥ 0.05)")

# ========================================
# 6. 추가 분석: 이탈률 분포 살펴보기
# ========================================
print(f"\n\n{'='*50}")
print("📌 이탈률 분포 상세 정보")
print(f"{'='*50}")

print("\n신규 방문자:")
print(f"  최소: {신규방문자['BounceRates'].min():.4f}")
print(f"  25%: {신규방문자['BounceRates'].quantile(0.25):.4f}")
print(f"  중앙값: {신규방문자['BounceRates'].median():.4f}")
print(f"  75%: {신규방문자['BounceRates'].quantile(0.75):.4f}")
print(f"  최대: {신규방문자['BounceRates'].max():.4f}")

print("\n재방문자:")
print(f"  최소: {재방문자['BounceRates'].min():.4f}")
print(f"  25%: {재방문자['BounceRates'].quantile(0.25):.4f}")
print(f"  중앙값: {재방문자['BounceRates'].median():.4f}")
print(f"  75%: {재방문자['BounceRates'].quantile(0.75):.4f}")
print(f"  최대: {재방문자['BounceRates'].max():.4f}")

# ========================================
# 7. 시각화
# ========================================
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 7-1. 평균 이탈률 막대그래프
ax1 = axes[0, 0]
그룹명 = ['신규 방문자', '재방문자']
평균값 = [신규_평균이탈률, 재방문_평균이탈률]
색상 = ['#e74c3c', '#27ae60']

bars = ax1.bar(그룹명, 평균값, color=색상, alpha=0.7, edgecolor='black', linewidth=2)
ax1.set_ylabel('평균 이탈률', fontweight='bold', fontsize=11)
ax1.set_title('신규 vs 재방문자 평균 이탈률 비교', fontweight='bold', fontsize=13)
ax1.set_ylim([0, max(평균값) * 1.3])

# 막대 위에 값 표시
for i, (bar, val) in enumerate(zip(bars, 평균값)):
    ax1.text(i, val + max(평균값)*0.02, f'{val:.4f}\n({val*100:.2f}%)', 
             ha='center', fontweight='bold', fontsize=10)

# 7-2. 박스플롯 (이탈률 분포)
ax2 = axes[0, 1]
data_to_plot = [신규방문자['BounceRates'], 재방문자['BounceRates']]
bp = ax2.boxplot(data_to_plot, labels=그룹명, patch_artist=True)

# 박스 색상 설정
for patch, color in zip(bp['boxes'], 색상):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax2.set_ylabel('이탈률', fontweight='bold', fontsize=11)
ax2.set_title('이탈률 분포 비교 (박스플롯)', fontweight='bold', fontsize=13)
ax2.grid(axis='y', alpha=0.3)

# 7-3. 히스토그램 (신규 방문자)
ax3 = axes[1, 0]
ax3.hist(신규방문자['BounceRates'], bins=30, color='#e74c3c', 
         alpha=0.7, edgecolor='black', label='신규 방문자')
ax3.axvline(신규_평균이탈률, color='red', linestyle='--', linewidth=2, 
            label=f'평균: {신규_평균이탈률:.4f}')
ax3.set_xlabel('이탈률', fontweight='bold')
ax3.set_ylabel('빈도', fontweight='bold')
ax3.set_title('신규 방문자 이탈률 분포', fontweight='bold', fontsize=13)
ax3.legend()
ax3.grid(axis='y', alpha=0.3)

# 7-4. 히스토그램 (재방문자)
ax4 = axes[1, 1]
ax4.hist(재방문자['BounceRates'], bins=30, color='#27ae60', 
         alpha=0.7, edgecolor='black', label='재방문자')
ax4.axvline(재방문_평균이탈률, color='green', linestyle='--', linewidth=2, 
            label=f'평균: {재방문_평균이탈률:.4f}')
ax4.set_xlabel('이탈률', fontweight='bold')
ax4.set_ylabel('빈도', fontweight='bold')
ax4.set_title('재방문자 이탈률 분포', fontweight='bold', fontsize=13)
ax4.legend()
ax4.grid(axis='y', alpha=0.3)

plt.tight_layout()
# plt.savefig('visitor_bounce_rate_comparison.png', dpi=300, bbox_inches='tight')
# print("\n\n📁 그래프가 'visitor_bounce_rate_comparison.png'로 저장되었습니다.")
plt.show()

# ========================================
# 8. 최종 결론 요약
# ========================================
print(f"\n\n{'='*60}")
print("🎯 최종 결론 요약")
print(f"{'='*60}")

print(f"\n1. 평균 이탈률:")
print(f"   • 신규 방문자: {신규_평균이탈률*100:.2f}%")
print(f"   • 재방문자: {재방문_평균이탈률*100:.2f}%")
print(f"   • 차이: {(신규_평균이탈률 - 재방문_평균이탈률)*100:.2f}%p")

print(f"\n2. 가설 검정:")
if 재방문_평균이탈률 < 신규_평균이탈률:
    print(f"   ✅ 재방문자의 이탈률이 더 낮습니다!")
    print(f"   → 가설이 입증되었습니다!")
else:
    print(f"   ❌ 가설이 기각되었습니다.")

print(f"\n3. 통계적 유의성:")
if p_value < 0.05:
    print(f"   ✅ 통계적으로 유의함 (p = {p_value:.6f})")
    print(f"   → 이 차이는 우연이 아닙니다!")
else:
    print(f"   ❌ 통계적으로 유의하지 않음 (p = {p_value:.6f})")

print(f"\n4. 실무 인사이트:")
print(f"   • 재방문자는 사이트에 익숙하여 이탈률이 낮습니다")
print(f"   • 신규 방문자 경험(UX) 개선이 중요합니다")
print(f"   • 재방문을 유도하는 마케팅 전략이 효과적입니다")

print(f"{'='*60}\n")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 한글 폰트 설정
plt.rcParams['font.family'] = 'appleGothic'
plt.rcParams['axes.unicode_minus'] = False

# ========================================
# 1. 데이터 불러오기
# ========================================
df = pd.read_csv('/Users/h/Desktop/ms/vscode/my/dataton_project/data/online_shoppers/online_shoppers_intention.csv')

print("📊 데이터 기본 정보")
print(f"전체 세션 수: {len(df):,}개")
print(f"신규 방문자: {(df['VisitorType'] == 'New_Visitor').sum():,}개")
print(f"재방문자: {(df['VisitorType'] == 'Returning_Visitor').sum():,}개")

# ========================================
# 2. 신규 vs 재방문자 구분하기
# ========================================
# 신규 방문자만 추출
신규방문자 = df[df['VisitorType'] == 'New_Visitor']

# 재방문자만 추출
재방문자 = df[df['VisitorType'] == 'Returning_Visitor']

print(f"\n✅ 그룹 분류 완료!")
print(f"- 신규 방문자: {len(신규방문자):,}명")
print(f"- 재방문자: {len(재방문자):,}명")

# ========================================
# 3. 평균 이탈률 계산하기
# ========================================
# 신규 방문자의 평균 이탈률
신규_평균이탈률 = 신규방문자['BounceRates'].mean()

# 재방문자의 평균 이탈률
재방문_평균이탈률 = 재방문자['BounceRates'].mean()

print(f"\n\n{'='*50}")
print("📈 평균 이탈률 비교")
print(f"{'='*50}")
print(f"신규 방문자 평균 이탈률: {신규_평균이탈률:.4f} ({신규_평균이탈률*100:.2f}%)")
print(f"재방문자 평균 이탈률:   {재방문_평균이탈률:.4f} ({재방문_평균이탈률*100:.2f}%)")
print(f"\n차이: {신규_평균이탈률 - 재방문_평균이탈률:.4f} ({(신규_평균이탈률 - 재방문_평균이탈률)*100:.2f}%p)")

# ========================================
# 4. 가설 검정하기
# ========================================
print(f"\n\n{'='*50}")
print("🔍 가설 검정 결과")
print(f"{'='*50}")

# 재방문자가 더 낮은지 확인
if 재방문_평균이탈률 < 신규_평균이탈률:
    차이_배율 = 신규_평균이탈률 / 재방문_평균이탈률
    print(f"✅ 가설 충족!")
    print(f"   재방문자의 이탈률이 더 낮습니다!")
    print(f"   → 신규 방문자 대비 {차이_배율:.2f}배 낮은 이탈률")
else:
    print(f"❌ 가설 불충족!")
    print(f"   신규 방문자의 이탈률이 더 낮습니다.")

# ========================================
# 5. 통계적 유의성 검정 (t-test)
# ========================================
# 두 그룹의 이탈률 차이가 통계적으로 의미있는지 확인
t_stat, p_value = stats.ttest_ind(신규방문자['BounceRates'], 
                                   재방문자['BounceRates'])

print(f"\n\n📊 통계적 유의성 검정")
print(f"t-통계량: {t_stat:.4f}")
print(f"p-value: {p_value:.6f}")

if p_value < 0.05:
    print("→ ✅ 통계적으로 유의한 차이가 있습니다! (p < 0.05)")
    print("   이 차이는 우연이 아닙니다!")
else:
    print("→ ❌ 통계적으로 유의한 차이가 없습니다. (p ≥ 0.05)")

# ========================================
# 6. 추가 분석: 이탈률 분포 살펴보기
# ========================================
print(f"\n\n{'='*50}")
print("📌 이탈률 분포 상세 정보")
print(f"{'='*50}")

print("\n신규 방문자:")
print(f"  최소: {신규방문자['BounceRates'].min():.4f}")
print(f"  25%: {신규방문자['BounceRates'].quantile(0.25):.4f}")
print(f"  중앙값: {신규방문자['BounceRates'].median():.4f}")
print(f"  75%: {신규방문자['BounceRates'].quantile(0.75):.4f}")
print(f"  최대: {신규방문자['BounceRates'].max():.4f}")

print("\n재방문자:")
print(f"  최소: {재방문자['BounceRates'].min():.4f}")
print(f"  25%: {재방문자['BounceRates'].quantile(0.25):.4f}")
print(f"  중앙값: {재방문자['BounceRates'].median():.4f}")
print(f"  75%: {재방문자['BounceRates'].quantile(0.75):.4f}")
print(f"  최대: {재방문자['BounceRates'].max():.4f}")

# ========================================
# 7. 시각화
# ========================================
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 7-1. 평균 이탈률 막대그래프
ax1 = axes[0, 0]
그룹명 = ['신규 방문자', '재방문자']
평균값 = [신규_평균이탈률, 재방문_평균이탈률]
색상 = ['#e74c3c', '#27ae60']

bars = ax1.bar(그룹명, 평균값, color=색상, alpha=0.7, edgecolor='black', linewidth=2)
ax1.set_ylabel('평균 이탈률', fontweight='bold', fontsize=11)
ax1.set_title('신규 vs 재방문자 평균 이탈률 비교', fontweight='bold', fontsize=13)
ax1.set_ylim([0, max(평균값) * 1.3])

# 막대 위에 값 표시
for i, (bar, val) in enumerate(zip(bars, 평균값)):
    ax1.text(i, val + max(평균값)*0.02, f'{val:.4f}\n({val*100:.2f}%)', 
             ha='center', fontweight='bold', fontsize=10)

# 7-2. 박스플롯 (이탈률 분포)
ax2 = axes[0, 1]
data_to_plot = [신규방문자['BounceRates'], 재방문자['BounceRates']]
bp = ax2.boxplot(data_to_plot, tick_labels=그룹명, patch_artist=True)

# 박스 색상 설정
for patch, color in zip(bp['boxes'], 색상):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax2.set_ylabel('이탈률', fontweight='bold', fontsize=11)
ax2.set_title('이탈률 분포 비교 (박스플롯)', fontweight='bold', fontsize=13)
ax2.grid(axis='y', alpha=0.3)

# 7-3. 히스토그램 (신규 방문자)
ax3 = axes[1, 0]
ax3.hist(신규방문자['BounceRates'], bins=30, color='#e74c3c', 
         alpha=0.7, edgecolor='black', label='신규 방문자')
ax3.axvline(신규_평균이탈률, color='red', linestyle='--', linewidth=2, 
            label=f'평균: {신규_평균이탈률:.4f}')
ax3.set_xlabel('이탈률', fontweight='bold')
ax3.set_ylabel('빈도', fontweight='bold')
ax3.set_title('신규 방문자 이탈률 분포', fontweight='bold', fontsize=13)
ax3.legend()
ax3.grid(axis='y', alpha=0.3)

# 7-4. 히스토그램 (재방문자)
ax4 = axes[1, 1]
ax4.hist(재방문자['BounceRates'], bins=30, color='#27ae60', 
         alpha=0.7, edgecolor='black', label='재방문자')
ax4.axvline(재방문_평균이탈률, color='green', linestyle='--', linewidth=2, 
            label=f'평균: {재방문_평균이탈률:.4f}')
ax4.set_xlabel('이탈률', fontweight='bold')
ax4.set_ylabel('빈도', fontweight='bold')
ax4.set_title('재방문자 이탈률 분포', fontweight='bold', fontsize=13)
ax4.legend()
ax4.grid(axis='y', alpha=0.3)

plt.tight_layout()
# plt.savefig('visitor_bounce_rate_comparison.png', dpi=300, bbox_inches='tight')
# print("\n\n📁 그래프가 'visitor_bounce_rate_comparison.png'로 저장되었습니다.")
plt.show()

# ========================================
# 8. 최종 결론 요약
# ========================================
print(f"\n\n{'='*60}")
print("🎯 최종 결론 요약")
print(f"{'='*60}")

print(f"\n1. 평균 이탈률:")
print(f"   • 신규 방문자: {신규_평균이탈률*100:.2f}%")
print(f"   • 재방문자: {재방문_평균이탈률*100:.2f}%")
print(f"   • 차이: {(신규_평균이탈률 - 재방문_평균이탈률)*100:.2f}%p")

print(f"\n2. 가설 검정:")
if 재방문_평균이탈률 < 신규_평균이탈률:
    print(f"   ✅ 재방문자의 이탈률이 더 낮습니다!")
    print(f"   → 가설이 입증되었습니다!")
else:
    print(f"   ❌ 가설이 기각되었습니다.")

print(f"\n3. 통계적 유의성:")
if p_value < 0.05:
    print(f"   ✅ 통계적으로 유의함 (p = {p_value:.6f})")
    print(f"   → 이 차이는 우연이 아닙니다!")
else:
    print(f"   ❌ 통계적으로 유의하지 않음 (p = {p_value:.6f})")

print(f"\n4. 실무 인사이트:")
print(f"   • 재방문자는 사이트에 익숙하여 이탈률이 낮습니다")
print(f"   • 신규 방문자 경험(UX) 개선이 중요합니다")
print(f"   • 재방문을 유도하는 마케팅 전략이 효과적입니다")

print(f"{'='*60}\n")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# 한글 폰트 설정
plt.rcParams['font.family'] = 'appleGothic'
plt.rcParams['axes.unicode_minus'] = False

# ========================================
# 1. 데이터 불러오기
# ========================================
df = pd.read_csv('/Users/h/Desktop/ms/vscode/my/dataton_project/data/online_shoppers/online_shoppers_intention.csv')

print("=" * 80)
print("🔍 재방문자 이탈 패턴: 로지스틱 회귀분석")
print("=" * 80)
print(f"\n전체 데이터: {len(df):,}건")

# ========================================
# 2. 재방문자 데이터 준비
# ========================================
재방문자 = df[df['VisitorType'] == 'Returning_Visitor'].copy()
print(f"재방문자: {len(재방문자):,}명")

# 총 페이지 수 계산
재방문자['총페이지수'] = (재방문자['Administrative'] + 
                      재방문자['Informational'] + 
                      재방문자['ProductRelated'])

# 총 체류시간 계산
재방문자['총체류시간'] = (재방문자['Administrative_Duration'] + 
                      재방문자['Informational_Duration'] + 
                      재방문자['ProductRelated_Duration'])

# 페이지당 체류시간 계산
재방문자['페이지당체류시간'] = 재방문자.apply(
    lambda x: x['총체류시간'] / x['총페이지수'] if x['총페이지수'] > 0 else 0, 
    axis=1
)

# 이탈 여부 (0: 이탈안함, 1: 이탈함)
재방문자['이탈'] = (재방문자['BounceRates'] > 0).astype(int)

print(f"\n이탈한 재방문자: {재방문자['이탈'].sum():,}명 ({재방문자['이탈'].mean()*100:.1f}%)")
print(f"이탈안한 재방문자: {(1-재방문자['이탈']).sum():,}명")

# ========================================
# 3. 로지스틱 회귀분석 준비
# ========================================
print("\n" + "=" * 80)
print("📊 로지스틱 회귀분석: 이탈에 영향을 미치는 요인")
print("=" * 80)

# 독립변수 선택
독립변수들 = ['총페이지수', '페이지당체류시간', '총체류시간', 'ExitRates', 'PageValues']

# 결측치 제거
분석데이터 = 재방문자[독립변수들 + ['이탈']].dropna()

print(f"\n분석 데이터: {len(분석데이터):,}건")
print(f"사용 변수: {', '.join(독립변수들)}")

# X, y 분리
X = 분석데이터[독립변수들]
y = 분석데이터['이탈']

# 표준화 (StandardScaler) - 계수 비교를 위해 필수!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=독립변수들)

print("\n✅ 데이터 표준화 완료 (평균=0, 표준편차=1)")

# ========================================
# 4. 로지스틱 회귀분석 실행
# ========================================
# 로지스틱 회귀 모델 학습
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_scaled, y)

# 계수 추출
계수 = pd.DataFrame({
    '변수': 독립변수들,
    '계수': model.coef_[0],
    '절편': model.intercept_[0]
})

# 계수의 절대값으로 중요도 계산
계수['중요도'] = abs(계수['계수'])
계수 = 계수.sort_values('중요도', ascending=False)

# Odds Ratio 계산 (해석하기 쉬운 지표)
계수['Odds_Ratio'] = np.exp(계수['계수'])

print("\n" + "=" * 80)
print("📈 로지스틱 회귀 계수 결과")
print("=" * 80)
print("\n계수 (Coefficient):")
print(계수[['변수', '계수', 'Odds_Ratio', '중요도']].to_string(index=False))

print(f"\n절편 (Intercept): {model.intercept_[0]:.4f}")
print(f"모델 정확도: {model.score(X_scaled, y)*100:.2f}%")

# ========================================
# 5. 계수 해석
# ========================================
print("\n" + "=" * 80)
print("💡 계수 해석 가이드")
print("=" * 80)

print("\n[계수(Coefficient)의 의미]")
print("• 양수(+): 값이 증가하면 이탈 확률이 증가")
print("• 음수(-): 값이 증가하면 이탈 확률이 감소")
print("• 절대값이 클수록 영향력이 큼")

print("\n[Odds Ratio의 의미]")
print("• > 1: 이탈 확률 증가 (예: 1.5 = 50% 증가)")
print("• < 1: 이탈 확률 감소 (예: 0.5 = 50% 감소)")
print("• = 1: 영향 없음")

print("\n" + "=" * 80)
print("🔍 주요 변수 상세 해석")
print("=" * 80)

for idx, row in 계수.head(3).iterrows():
    변수명 = row['변수']
    계수값 = row['계수']
    odds = row['Odds_Ratio']
    
    print(f"\n📌 {변수명}:")
    print(f"   계수: {계수값:+.4f}")
    print(f"   Odds Ratio: {odds:.4f}")
    
    if 계수값 > 0:
        변화율 = (odds - 1) * 100
        print(f"   → {변수명}이 1 표준편차 증가하면 이탈 확률이 {변화율:+.1f}% 변화")
        print(f"   💡 해석: {변수명}이 높을수록 이탈 가능성 증가!")
    else:
        변화율 = (1 - odds) * 100
        print(f"   → {변수명}이 1 표준편차 증가하면 이탈 확률이 {변화율:+.1f}% 감소")
        print(f"   💡 해석: {변수명}이 높을수록 이탈 가능성 감소!")

# ========================================
# 6. 가설 검정: 페이지당 체류시간의 영향
# ========================================
print("\n" + "=" * 80)
print("🎯 가설 검정: 페이지당 체류시간이 이탈에 미치는 영향")
print("=" * 80)

페이지당시간_계수 = 계수[계수['변수'] == '페이지당체류시간']['계수'].values[0]
페이지당시간_odds = 계수[계수['변수'] == '페이지당체류시간']['Odds_Ratio'].values[0]

print(f"\n페이지당체류시간 계수: {페이지당시간_계수:+.4f}")
print(f"Odds Ratio: {페이지당시간_odds:.4f}")

if 페이지당시간_계수 < 0:
    print("\n✅ 가설 충족!")
    print(f"   페이지당 체류시간이 짧을수록 이탈 확률이 증가합니다")
    변화율 = (1 - 페이지당시간_odds) * 100
    print(f"   → 페이지당 체류시간이 1 표준편차 증가하면 이탈 확률이 {변화율:.1f}% 감소")
    print(f"\n💡 이탈한 재방문자는 페이지를 빠르게 훑어보는 패턴!")
else:
    print("\n❌ 가설 불충족")

# ========================================
# 7. 기술통계 비교
# ========================================
print("\n" + "=" * 80)
print("📊 이탈 여부별 기술통계")
print("=" * 80)

이탈그룹 = 분석데이터[분석데이터['이탈'] == 1]
비이탈그룹 = 분석데이터[분석데이터['이탈'] == 0]

print("\n주요 변수 평균 비교:")
비교표 = pd.DataFrame({
    '변수': 독립변수들,
    '이탈함': [이탈그룹[v].mean() for v in 독립변수들],
    '이탈안함': [비이탈그룹[v].mean() for v in 독립변수들],
})
비교표['차이'] = 비교표['이탈함'] - 비교표['이탈안함']
비교표['차이율(%)'] = (비교표['차이'] / 비교표['이탈안함'] * 100).round(1)

print(비교표.to_string(index=False))

# ========================================
# 8. 시각화
# ========================================
fig = plt.figure(figsize=(16, 10))

# 8-1. 회귀 계수 막대그래프
ax1 = plt.subplot(2, 3, 1)
계수_정렬 = 계수.sort_values('계수')
colors = ['#e74c3c' if x > 0 else '#27ae60' for x in 계수_정렬['계수']]
ax1.barh(계수_정렬['변수'], 계수_정렬['계수'], color=colors, alpha=0.7, edgecolor='black')
ax1.axvline(x=0, color='black', linestyle='-', linewidth=1)
ax1.set_xlabel('계수 (Coefficient)', fontweight='bold')
ax1.set_title('로지스틱 회귀 계수 (이탈 예측)', fontweight='bold', fontsize=13)
for i, (var, coef) in enumerate(zip(계수_정렬['변수'], 계수_정렬['계수'])):
    ax1.text(coef + (0.02 if coef > 0 else -0.02), i, f'{coef:+.3f}', 
             va='center', ha='left' if coef > 0 else 'right', fontweight='bold', fontsize=9)
ax1.grid(axis='x', alpha=0.3)

# 8-2. Odds Ratio
ax2 = plt.subplot(2, 3, 2)
계수_정렬_odds = 계수.sort_values('Odds_Ratio')
colors_odds = ['#e74c3c' if x > 1 else '#27ae60' for x in 계수_정렬_odds['Odds_Ratio']]
ax2.barh(계수_정렬_odds['변수'], 계수_정렬_odds['Odds_Ratio'], 
         color=colors_odds, alpha=0.7, edgecolor='black')
ax2.axvline(x=1, color='black', linestyle='--', linewidth=2, label='기준선 (OR=1)')
ax2.set_xlabel('Odds Ratio', fontweight='bold')
ax2.set_title('Odds Ratio (이탈 위험도)', fontweight='bold', fontsize=13)
ax2.legend()
for i, (var, odds) in enumerate(zip(계수_정렬_odds['변수'], 계수_정렬_odds['Odds_Ratio'])):
    ax2.text(odds + 0.02, i, f'{odds:.3f}', va='center', fontweight='bold', fontsize=9)
ax2.grid(axis='x', alpha=0.3)

# 8-3. 변수 중요도
ax3 = plt.subplot(2, 3, 3)
계수_중요도순 = 계수.sort_values('중요도', ascending=True)
ax3.barh(계수_중요도순['변수'], 계수_중요도순['중요도'], 
         color='#3498db', alpha=0.7, edgecolor='black')
ax3.set_xlabel('중요도 (|계수|)', fontweight='bold')
ax3.set_title('변수 중요도 순위', fontweight='bold', fontsize=13)
for i, (var, imp) in enumerate(zip(계수_중요도순['변수'], 계수_중요도순['중요도'])):
    ax3.text(imp + 0.01, i, f'{imp:.3f}', va='center', fontweight='bold', fontsize=9)
ax3.grid(axis='x', alpha=0.3)

# 8-4. 페이지당 체류시간 비교
ax4 = plt.subplot(2, 3, 4)
그룹명 = ['이탈함', '이탈안함']
시간평균 = [이탈그룹['페이지당체류시간'].mean(), 비이탈그룹['페이지당체류시간'].mean()]
colors_bar = ['#e74c3c', '#27ae60']
bars = ax4.bar(그룹명, 시간평균, color=colors_bar, alpha=0.7, edgecolor='black', linewidth=2)
ax4.set_ylabel('페이지당 체류시간 (초)', fontweight='bold')
ax4.set_title('페이지당 체류시간 비교', fontweight='bold', fontsize=13)
for i, (bar, val) in enumerate(zip(bars, 시간평균)):
    ax4.text(i, val + 5, f'{val:.2f}초', ha='center', fontweight='bold', fontsize=11)
ax4.grid(axis='y', alpha=0.3)

# 8-5. 총 페이지 수 비교
ax5 = plt.subplot(2, 3, 5)
페이지평균 = [이탈그룹['총페이지수'].mean(), 비이탈그룹['총페이지수'].mean()]
bars = ax5.bar(그룹명, 페이지평균, color=colors_bar, alpha=0.7, edgecolor='black', linewidth=2)
ax5.set_ylabel('총 페이지 수', fontweight='bold')
ax5.set_title('총 페이지 수 비교', fontweight='bold', fontsize=13)
for i, (bar, val) in enumerate(zip(bars, 페이지평균)):
    ax5.text(i, val + 0.5, f'{val:.2f}개', ha='center', fontweight='bold', fontsize=11)
ax5.grid(axis='y', alpha=0.3)

# 8-6. PageValues 비교
ax6 = plt.subplot(2, 3, 6)
pv평균 = [이탈그룹['PageValues'].mean(), 비이탈그룹['PageValues'].mean()]
bars = ax6.bar(그룹명, pv평균, color=colors_bar, alpha=0.7, edgecolor='black', linewidth=2)
ax6.set_ylabel('PageValues', fontweight='bold')
ax6.set_title('PageValues 비교', fontweight='bold', fontsize=13)
for i, (bar, val) in enumerate(zip(bars, pv평균)):
    ax6.text(i, val + 1, f'{val:.2f}', ha='center', fontweight='bold', fontsize=11)
ax6.grid(axis='y', alpha=0.3)

plt.tight_layout()
# plt.savefig('logistic_regression_bounce_analysis.png', dpi=300, bbox_inches='tight')
# print("\n📊 그래프가 'logistic_regression_bounce_analysis.png'로 저장되었습니다!")
plt.show()

# ========================================
# 9. 최종 결론
# ========================================
print("\n" + "=" * 80)
print("🎯 최종 결론 및 마케팅 제안")
print("=" * 80)

print("\n1️⃣ 이탈에 가장 큰 영향을 미치는 요인 TOP 3:")
for i, row in 계수.head(3).iterrows():
    print(f"   {row['변수']}: 계수 {row['계수']:+.4f} (중요도: {row['중요도']:.4f})")

print(f"\n2️⃣ 페이지당 체류시간의 영향:")
if 페이지당시간_계수 < 0:
    print(f"   ✅ 페이지당 체류시간이 짧을수록 이탈 확률 증가")
    print(f"   → 이탈한 재방문자는 페이지를 빠르게 넘기는 패턴!")
    print(f"   → 계수: {페이지당시간_계수:+.4f}")

print(f"\n3️⃣ 실무 적용 인사이트:")
print(f"   • 재방문자는 '목적 지향적'으로 쇼핑합니다")
print(f"   • 빠르게 훑어보고 관심 없으면 즉시 이탈합니다")
print(f"   • 첫 페이지에서 개인화된 추천이 매우 중요합니다")

print(f"\n4️⃣ 마케팅 전략:")
print(f"   📌 첫 페이지 최적화:")
print(f"      - 신상품을 상단에 배치")
print(f"      - 개인화 추천 알고리즘 강화")
print(f"      - '당신을 위한 추천' 섹션 추가")
print(f"   📌 체류시간 증가 전략:")
print(f"      - 흥미로운 콘텐츠/이벤트 배너")
print(f"      - 한정 할인/타임딜 알림")
print(f"      - 위시리스트 알림 기능")

print("=" * 80)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 한글 폰트 설정
plt.rcParams['font.family'] = 'appleGothic'
plt.rcParams['axes.unicode_minus'] = False

# ========================================
# 1. 데이터 불러오기
# ========================================
df = pd.read_csv('/Users/h/Desktop/ms/vscode/my/dataton_project/data/online_shoppers/online_shoppers_intention.csv')

print("=" * 80)
print("🔍 재방문자 이탈 행동 패턴 분석")
print("=" * 80)
print(f"\n전체 데이터: {len(df):,}건")

# ========================================
# 2. 필요한 데이터만 추출하기
# ========================================
# 재방문자만 선택
재방문자 = df[df['VisitorType'] == 'Returning_Visitor'].copy()

print(f"재방문자: {len(재방문자):,}명")

# 전체 페이지 수 계산 (Administrative + Informational + ProductRelated)
재방문자['총페이지수'] = (재방문자['Administrative'] + 
                      재방문자['Informational'] + 
                      재방문자['ProductRelated'])

# 전체 체류시간 계산
재방문자['총체류시간'] = (재방문자['Administrative_Duration'] + 
                      재방문자['Informational_Duration'] + 
                      재방문자['ProductRelated_Duration'])

# 페이지당 평균 체류시간 계산 (0으로 나누기 방지)
재방문자['페이지당체류시간'] = 재방문자.apply(
    lambda x: x['총체류시간'] / x['총페이지수'] if x['총페이지수'] > 0 else 0, 
    axis=1
)

# ========================================
# 3. 이탈 여부로 그룹 나누기
# ========================================
# BounceRates > 0 이면 이탈한 것으로 간주
재방문자['이탈여부'] = 재방문자['BounceRates'].apply(
    lambda x: '이탈함' if x > 0 else '이탈안함'
)

이탈그룹 = 재방문자[재방문자['이탈여부'] == '이탈함']
비이탈그룹 = 재방문자[재방문자['이탈여부'] == '이탈안함']

print(f"\n이탈한 재방문자: {len(이탈그룹):,}명")
print(f"이탈안한 재방문자: {len(비이탈그룹):,}명")

# ========================================
# 4. 가설 1: 이탈한 재방문자는 여러 페이지를 본다
# ========================================
print("\n" + "=" * 80)
print("📄 가설 1: 이탈한 재방문자는 여러 페이지를 볼까?")
print("=" * 80)

이탈_평균페이지 = 이탈그룹['총페이지수'].mean()
비이탈_평균페이지 = 비이탈그룹['총페이지수'].mean()

print(f"\n이탈한 그룹 평균 페이지 수: {이탈_평균페이지:.2f}개")
print(f"이탈안한 그룹 평균 페이지 수: {비이탈_평균페이지:.2f}개")
print(f"차이: {이탈_평균페이지 - 비이탈_평균페이지:+.2f}개")

# t-검정으로 통계적 유의성 확인
t_stat_page, p_value_page = stats.ttest_ind(
    이탈그룹['총페이지수'].dropna(),
    비이탈그룹['총페이지수'].dropna()
)

print(f"\nt-통계량: {t_stat_page:.4f}")
print(f"p-value: {p_value_page:.6f}")

if 이탈_평균페이지 > 비이탈_평균페이지:
    print("\n✅ 가설 1 부분 충족: 이탈한 그룹이 더 많은 페이지를 봅니다")
    if p_value_page < 0.05:
        print("   → 통계적으로 유의함! (p < 0.05)")
    else:
        print("   → 통계적으로 유의하지 않음 (p ≥ 0.05)")
else:
    print("\n❌ 가설 1 불충족: 이탈안한 그룹이 더 많은 페이지를 봅니다")

# ========================================
# 5. 가설 2: 페이지당 체류시간이 적을 것이다
# ========================================
print("\n" + "=" * 80)
print("⏱️ 가설 2: 이탈한 재방문자는 페이지당 체류시간이 짧을까?")
print("=" * 80)

이탈_페이지당시간 = 이탈그룹['페이지당체류시간'].mean()
비이탈_페이지당시간 = 비이탈그룹['페이지당체류시간'].mean()

print(f"\n이탈한 그룹 페이지당 체류시간: {이탈_페이지당시간:.2f}초")
print(f"이탈안한 그룹 페이지당 체류시간: {비이탈_페이지당시간:.2f}초")
print(f"차이: {이탈_페이지당시간 - 비이탈_페이지당시간:+.2f}초")

# t-검정으로 통계적 유의성 확인
t_stat_time, p_value_time = stats.ttest_ind(
    이탈그룹['페이지당체류시간'].dropna(),
    비이탈그룹['페이지당체류시간'].dropna()
)

print(f"\nt-통계량: {t_stat_time:.4f}")
print(f"p-value: {p_value_time:.6f}")

if 이탈_페이지당시간 < 비이탈_페이지당시간:
    print("\n✅ 가설 2 충족: 이탈한 그룹의 페이지당 체류시간이 더 짧습니다!")
    if p_value_time < 0.05:
        print("   → 통계적으로 유의함! (p < 0.05)")
    else:
        print("   → 통계적으로 유의하지 않음 (p ≥ 0.05)")
else:
    print("\n❌ 가설 2 불충족: 이탈한 그룹의 페이지당 체류시간이 더 깁니다")

# ========================================
# 6. 추가 분석: 총 체류시간 비교
# ========================================
print("\n" + "=" * 80)
print("⏰ 추가 분석: 총 체류시간 비교")
print("=" * 80)

이탈_총시간 = 이탈그룹['총체류시간'].mean()
비이탈_총시간 = 비이탈그룹['총체류시간'].mean()

print(f"\n이탈한 그룹 총 체류시간: {이탈_총시간:.2f}초")
print(f"이탈안한 그룹 총 체류시간: {비이탈_총시간:.2f}초")
print(f"차이: {이탈_총시간 - 비이탈_총시간:+.2f}초")

# ========================================
# 7. 상세 분석: 분포 확인
# ========================================
print("\n" + "=" * 80)
print("📊 상세 분포 정보")
print("=" * 80)

print("\n[페이지 수 분포]")
print("이탈한 그룹:")
print(f"  최소: {이탈그룹['총페이지수'].min():.0f}개")
print(f"  중앙값: {이탈그룹['총페이지수'].median():.0f}개")
print(f"  최대: {이탈그룹['총페이지수'].max():.0f}개")

print("\n이탈안한 그룹:")
print(f"  최소: {비이탈그룹['총페이지수'].min():.0f}개")
print(f"  중앙값: {비이탈그룹['총페이지수'].median():.0f}개")
print(f"  최대: {비이탈그룹['총페이지수'].max():.0f}개")

print("\n[페이지당 체류시간 분포]")
print("이탈한 그룹:")
print(f"  최소: {이탈그룹['페이지당체류시간'].min():.2f}초")
print(f"  중앙값: {이탈그룹['페이지당체류시간'].median():.2f}초")
print(f"  최대: {이탈그룹['페이지당체류시간'].max():.2f}초")

print("\n이탈안한 그룹:")
print(f"  최소: {비이탈그룹['페이지당체류시간'].min():.2f}초")
print(f"  중앙값: {비이탈그룹['페이지당체류시간'].median():.2f}초")
print(f"  최대: {비이탈그룹['페이지당체류시간'].max():.2f}초")

# ========================================
# 8. 시각화
# ========================================
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('재방문자 이탈 행동 패턴 분석', fontsize=16, fontweight='bold')

# 8-1. 평균 페이지 수 비교
ax1 = axes[0, 0]
그룹명 = ['이탈함', '이탈안함']
페이지평균 = [이탈_평균페이지, 비이탈_평균페이지]
colors = ['#e74c3c', '#27ae60']

bars = ax1.bar(그룹명, 페이지평균, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax1.set_ylabel('평균 페이지 수', fontweight='bold')
ax1.set_title('평균 페이지 수 비교', fontweight='bold', fontsize=13)
for i, (bar, val) in enumerate(zip(bars, 페이지평균)):
    ax1.text(i, val + 0.5, f'{val:.2f}개', ha='center', fontweight='bold', fontsize=11)
ax1.grid(axis='y', alpha=0.3)

# 8-2. 페이지당 체류시간 비교
ax2 = axes[0, 1]
시간평균 = [이탈_페이지당시간, 비이탈_페이지당시간]

bars = ax2.bar(그룹명, 시간평균, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax2.set_ylabel('페이지당 체류시간 (초)', fontweight='bold')
ax2.set_title('페이지당 평균 체류시간 비교', fontweight='bold', fontsize=13)
for i, (bar, val) in enumerate(zip(bars, 시간평균)):
    ax2.text(i, val + 5, f'{val:.2f}초', ha='center', fontweight='bold', fontsize=11)
ax2.grid(axis='y', alpha=0.3)

# 8-3. 총 체류시간 비교
ax3 = axes[0, 2]
총시간평균 = [이탈_총시간, 비이탈_총시간]

bars = ax3.bar(그룹명, 총시간평균, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax3.set_ylabel('총 체류시간 (초)', fontweight='bold')
ax3.set_title('총 체류시간 비교', fontweight='bold', fontsize=13)
for i, (bar, val) in enumerate(zip(bars, 총시간평균)):
    ax3.text(i, val + 20, f'{val:.0f}초', ha='center', fontweight='bold', fontsize=11)
ax3.grid(axis='y', alpha=0.3)

# 8-4. 페이지 수 박스플롯
ax4 = axes[1, 0]
data_page = [이탈그룹['총페이지수'], 비이탈그룹['총페이지수']]
bp1 = ax4.boxplot(data_page, tick_labels=그룹명, patch_artist=True)
for patch, color in zip(bp1['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
ax4.set_ylabel('페이지 수', fontweight='bold')
ax4.set_title('페이지 수 분포 (박스플롯)', fontweight='bold', fontsize=13)
ax4.grid(axis='y', alpha=0.3)

# 8-5. 페이지당 체류시간 박스플롯
ax5 = axes[1, 1]
data_time = [이탈그룹['페이지당체류시간'], 비이탈그룹['페이지당체류시간']]
bp2 = ax5.boxplot(data_time, tick_labels=그룹명, patch_artist=True)
for patch, color in zip(bp2['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
ax5.set_ylabel('페이지당 체류시간 (초)', fontweight='bold')
ax5.set_title('페이지당 체류시간 분포 (박스플롯)', fontweight='bold', fontsize=13)
ax5.grid(axis='y', alpha=0.3)

# 8-6. 산점도: 페이지 수 vs 페이지당 체류시간
ax6 = axes[1, 2]
ax6.scatter(이탈그룹['총페이지수'], 이탈그룹['페이지당체류시간'], 
           alpha=0.3, c='#e74c3c', label='이탈함', s=30)
ax6.scatter(비이탈그룹['총페이지수'], 비이탈그룹['페이지당체류시간'], 
           alpha=0.3, c='#27ae60', label='이탈안함', s=30)
ax6.set_xlabel('총 페이지 수', fontweight='bold')
ax6.set_ylabel('페이지당 체류시간 (초)', fontweight='bold')
ax6.set_title('페이지 수 vs 페이지당 체류시간', fontweight='bold', fontsize=13)
ax6.legend()
ax6.grid(alpha=0.3)

plt.tight_layout()
# plt.savefig('returning_visitor_bounce_analysis.png', dpi=300, bbox_inches='tight')
# print("\n📊 그래프가 'returning_visitor_bounce_analysis.png'로 저장되었습니다!")
plt.show()

# ========================================
# 9. 최종 결론
# ========================================
print("\n" + "=" * 80)
print("🎯 최종 결론")
print("=" * 80)

print(f"\n1️⃣ 페이지 수:")
if 이탈_평균페이지 > 비이탈_평균페이지:
    print(f"   ✅ 이탈한 그룹이 {이탈_평균페이지 - 비이탈_평균페이지:.2f}개 더 많은 페이지를 봅니다")
else:
    print(f"   ❌ 이탈안한 그룹이 {비이탈_평균페이지 - 이탈_평균페이지:.2f}개 더 많은 페이지를 봅니다")

print(f"\n2️⃣ 페이지당 체류시간:")
if 이탈_페이지당시간 < 비이탈_페이지당시간:
    print(f"   ✅ 이탈한 그룹이 페이지당 {비이탈_페이지당시간 - 이탈_페이지당시간:.2f}초 더 짧게 머뭅니다!")
    차이비율 = (비이탈_페이지당시간 - 이탈_페이지당시간) / 비이탈_페이지당시간 * 100
    print(f"   → 약 {차이비율:.1f}% 더 빠르게 페이지를 넘깁니다")
else:
    print(f"   ❌ 이탈한 그룹이 페이지당 {이탈_페이지당시간 - 비이탈_페이지당시간:.2f}초 더 길게 머뭅니다")

print(f"\n3️⃣ 통계적 유의성:")
print(f"   페이지 수: p = {p_value_page:.6f} {'✅ 유의함' if p_value_page < 0.05 else '❌ 유의하지 않음'}")
print(f"   페이지당 체류시간: p = {p_value_time:.6f} {'✅ 유의함' if p_value_time < 0.05 else '❌ 유의하지 않음'}")

print(f"\n💡 해석:")
if 이탈_페이지당시간 < 비이탈_페이지당시간:
    print("   • 이탈한 재방문자는 페이지를 '빠르게 훑어보는' 패턴을 보입니다")
    print("   • 원하는 상품을 찾지 못하면 빠르게 다음 페이지로 넘어갑니다")
    print("   • '목적 지향적'이고 '효율적인' 쇼핑 행동입니다")
    print("\n📌 마케팅 제안:")
    print("   → 첫 페이지에서 개인화된 추천 상품을 보여주세요")
    print("   → 신상품, 할인 상품을 눈에 띄게 배치하세요")
    print("   → 재방문자가 빠르게 원하는 것을 찾도록 검색 기능을 강화하세요")

print("=" * 80)
```
